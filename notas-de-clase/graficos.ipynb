## Graficando funciones

```python
x = np.linspace(0, 2 * np.pi, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
x
```

```python
plt.plot(x, np.sin(x), linestyle="-", label="sin(x)")
plt.plot(x, np.sin(2 * x), linestyle="--", label="sin(2x)")
plt.plot(x, np.sin(3 * x), linestyle="-.", label="sin(3x)")
plt.legend()
plt.show()
```
